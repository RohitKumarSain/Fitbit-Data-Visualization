EDA: Insights from the data

```python
# Lets find how many days of this data?
date.size
```

```python
# What is the average step count?
# np.average(step_count)
print("Average step count:",step_count.mean())
```

```python
#  On which day the step count was highest?
print(f"Highest step count is {step_count.max()} on {date[step_count.argmax()]}")
```

```python
# On which day the step count was minimum?
print(f"Highest step count is {step_count.min()} on {date[step_count.argmin()]}")
```

```python
# Lets check the calories burnt on the highest step count day?
print(f"{calories_burned.max()} calories burnt on highest step count on {date[step_count.argmax()]}")
```

```python
# Lets check the calories burnt on the minimum step count day?
print(f"{calories_burned[step_count.argmin()]} calories burnt on highest step count on {date[step_count.argmin()]}")
```

```python
# What is the most frequent mood?
for i in range(np.unique(mood).size):
    mood_keys = np.unique(mood,return_counts=True)[0]
    mood_values = np.unique(mood,return_counts=True)[1]
    print(mood_keys[i],"=",mood_values[i])

plt.pie(mood_values,labels=mood_keys,autopct="%1.1f%%")
plt.show()
```

```python
# Comparing step counts on bad mood days and good mood days.

good_mood_days = date[mood=="Happy"].size
bad_mood_days = date[mood=="Sad"].size
print("Total good mood days are",good_mood_days)
print("Total bad mood days are",bad_mood_days)

avg_step_count_on_good_days = step_count[mood=="Happy"].sum()/step_count[mood=="Happy"].size
avg_step_count_on_bad_days = step_count[mood=="Sad"].sum()/step_count[mood=="Sad"].size

print("Average Step Counts on Good Days = ",int(avg_step_count_on_good_days))
print("Average Step Counts on Bad Days = ",int(avg_step_count_on_bad_days))
```

```python
# How many Highest / Lowest calories burned on which day?

print(f"Highest Calories {calories_burned.max()} on {date[calories_burned.argmax()]}")
print(f"Lowest Calories {calories_burned.min()} on {date[calories_burned.argmin()]}")

plt.plot(date,calories_burned,marker="o",color="blue",linewidth=2)
plt.title("Calories Burned")
plt.xlabel("dates")
plt.ylabel("Calories Burned")
plt.xticks(rotation=90)
plt.tight_layout(rect=[0,0,3,2])
plt.show()
```

```python
#  Activness: How much active or inactive
total_active_days= activeness[activeness=="Active"].size
total_inactive_days= activeness[activeness=="Inactive"].size
print(f"Total Active days {total_active_days} in {date.size} days")
print(f"Total Inactive days {total_inactive_days} in {date.size} days")

activeness_keys = np.unique(activeness,return_counts=True)[0]
activeness_values = np.unique(activeness,return_counts=True)[1]
print(activeness_keys,activeness_values)
plt.pie(activeness_values,labels=activeness_keys,autopct="%1.1f%%")
plt.title("Active and Inactive Percentage")
plt.show()
```

```python
# How much calories burned for 1 kg weight
unique_weight = np.unique(weight)
weightList=[]
calories_burnedList=[]

for w in unique_weight:
    print("Total Calories Burned:",calories_burned[weight==w].sum(),"for reducing weight",w,"kg")
    weightList.append(w)
    calories_burnedList.append(calories_burned[weight==w].sum())

plt.pie(calories_burnedList,labels=weightList,startangle=90,autopct=lambda p : '{:.0f}'.format(p*sum(calories_burnedList)/100))
plt.title("How much calories burned for per kg weight")
plt.legend()
plt.show()
```
